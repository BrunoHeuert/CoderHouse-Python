Exercício de Visualização

Crie visualizações usando o arquivo fastfood.xlsx, como:

1) Gráfico de barras
2) Histograma
3) Distribuição
4) Dispersão

```python
#importa as bilbiotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#lê a planilha num dataframe
df = pd.read_excel("fastfood.xlsx")
df.head()
```

```python
#instala o pip do seaborn
#pip install seaborn
```

```python
#cria outro dataframe que recebe o df agrupado pelos restaurantes, trazendo todas as médias numéricas
fastfood = df.groupby("restaurant").mean(numeric_only=True).reset_index()
fastfood.head()
```

```python
#Gráfico de barras
#o plt.bar define qual coluna será o eixo x e qual será o eixo y e define a cor das barras
plt.bar(fastfood["restaurant"], fastfood["calories"], color="red")
#o plt.xticks mostra a rotação do rotulo das colunas num angulo de 45 graus
plt.xticks(rotation=45)
#plt.title define o titulo
plt.title("Média de Calorias por Restaurante")
#define o titulo do eixo x
plt.xlabel("Restaurantes")
#define o titulo do eixo y
plt.ylabel("Calorias")
#mostra o gráfico
plt.show()
```

```python
#Histograma
#define qual coluna será usada
plt.hist(fastfood["calories"], bins=4)
#define um titulo
plt.title("Histograma de Calorias")
#mostra o gráfico
plt.show()
```

```python
#Distribuição
#define qual coluna e gráfico será usado
sns.displot(fastfood["calories"], kde=True)
plt.show()
```

```python
#Dispersão
#Define a coluna e o tipo de gráfico
sns.scatterplot(x="calories", y="sugar", data=fastfood, size="protein", alpha=0.7)
plt.show()
```
