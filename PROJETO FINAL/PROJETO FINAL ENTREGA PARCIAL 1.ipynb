```python
#Importa as bilbiotecas
import pandas as pd
import requests
```

```python
#Realiza o get da API e atribui a uma variável a sua resposta
url = 'https://pokeapi.co/api/v2/pokemon'
resposta = requests.get(url)
print(resposta)
```

```python
#Se o status_code da resposta for 200, ele atribui os valores dela dentro da variável data_json
if resposta.status_code == 200:
    datapoke_json = resposta.json()
#Caso o status_code da API for diferente de 200, ele aplica o alerta de erro
else:
    from plyer import notification
    notification.notify(
        title= 'ERRO AO CONSULTAR A API',
        message='A API não está retornando um valor para acesso.',
        app_name='Python',
        timeout=10
    )
```

```python
#Mostra o json da API
datapoke_json
```

```python
#Declara uma variável na qual recebe os valores do JSON filtrando pela chave "results", pela primeira posição da lista e em seguida faz um get do JSON do endpoint
get_pokemon = requests.get(datapoke_json['results'][0]['url']).json()
get_pokemon
```

```python
#cria a primeira tebela, ela vai ser uma tabela de ligação, possuindo apenas o id e nome do pokemon
pokemon = pd.DataFrame({
    "id_pokemon": [get_pokemon['order']],
    "nome_pokemon": [get_pokemon['name']]
})
pokemon.head()
```

```python
#cria a segunda tabela, essa é responsável por mostrar os status_base do pokemon
pokemon_status_base = pd.DataFrame({
    "id_pokemon": [get_pokemon['order']],
    "hp": [get_pokemon['stats'][0]['base_stat']],
    "ataque": [get_pokemon['stats'][1]['base_stat']],
    "defesa": [get_pokemon['stats'][2]['base_stat']],
    "ataque_especial": [get_pokemon['stats'][3]['base_stat']],
    "defesa_especial": [get_pokemon['stats'][4]['base_stat']],
    "velocidade": [get_pokemon['stats'][5]['base_stat']]
})
pokemon_status_base.head()
```

```python
#cria a terceira tabela, na qual informa a tipagem do pokemon
pokemon_tipo = pd.DataFrame({
    "id_pokemon": [get_pokemon['order']],
    "tipo_primario": [get_pokemon['types'][0]['type']['name']],
    "tipo_secundario": [get_pokemon['types'][1]['type']['name']]
})
pokemon_tipo.head()
```

```python
#cria a quarta tabela, que msotra as habilidades do pokemon
pokemon_habilidades = pd.DataFrame({
    "id_pokemon": [get_pokemon['order']],
    "habilidade_1": [get_pokemon['abilities'][0]['ability']['name']],
    "habilidade_2": [get_pokemon['abilities'][1]['ability']['name']]
})
pokemon_habilidades.head()
```

```python
#por fim, cria a quinta tabela, contendo algumas informações gerais do pokemon
pokemon_inf_gerais = pd.DataFrame({
    "id_pokemon,": [get_pokemon['order']],
    "base_de_xp": [get_pokemon['base_experience']],
    "altura": [get_pokemon['height']],
    "peso": [get_pokemon['weight']],
    "especie": [get_pokemon['species']['name']]
})
pokemon_inf_gerais.head()
```
