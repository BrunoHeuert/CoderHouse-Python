1) abra o arquivo "Cancer_Data.csv", conte quantas linhas o dataframe possui com a coluna diagnosis = 'B'

```python
import pandas as pd
df = pd.read_csv('Cancer_Data.csv',#"importa" a planilha csv
                 sep = ',', #define o separador de colunas
                 decimal = '.') #define o separador das casas decimais
df.head()#le os 5 primeiro resultados da planilha
```

```python
df.loc[df['diagnosis'] == 'B', 'diagnosis'].count()
#dentro do .loc são parametrizados dois parametros 
#o primeiro "['diagnosis'] == 'B'" vai especificar quais linhas devem ser acessadas, nesse caso, todas que forem ''diagnosis' == 'B'
#o segundo "'diagnosis'" vai especificar a coluna acessada, nesse caso a própria diagnosis
#note, primeiro ele vai acessar as linhas que tal linha seja tal valor, e depois ele filtra a coluna acessada
```

2) Substitua a coluna diagnosis seguinte a regra:

    A. "B" para "Benign cancer"

    B. "M" para "Malignant cancer"

```python
df['diagnosis'] = df['diagnosis'].str.replace('M', 'malignant cancer').replace('B', 'benign cancer')
#a coluna dianosis do df recebe o valor substituido
df
```
